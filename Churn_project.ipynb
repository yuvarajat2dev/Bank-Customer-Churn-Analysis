```python
import pandas as pd
import numpy as np
```

Project Goal

Analyze customer churn and identify factors associated with customer retention and attrition.

```python
df=pd.read_csv('Churn_Modelling.csv')
```

Customer Churn Overview

Q1. What is the overall churn rate?

```python
(df['Exited'].mean() *100 ).round(2)
```

The dataset reveals an overall churn rate of 20.37%, meaning that roughly one-fifth of the customer base has left the bank. Understanding the characteristics of these customers can help identify opportunities to improve retention and reduce future churn.

Q2. How many customers stayed and how many left?

```python
df['Exited'].value_counts()
```

Demographic Analysis

Q3. Which country has the highest churn rate?

```python
(df.groupby('Geography')['Exited'].mean()*100).sort_values(ascending=False).round(2)
```

Germany has the highest churn rate compared to other countries, indicating that customers from this region are more likely to leave the bank. In contrast, France and Spain show relatively lower churn rates, suggesting better customer retention in these markets.

Q4. Is churn higher among males or females?

```python
df.groupby('Gender')['Exited'].mean()*100
```

The analysis shows that churn is higher among female customers compared to male customers. This suggests that female customers are more likely to exit the bank, indicating a potential need to investigate service experience, product fit, or engagement strategies targeted toward this segment.

Q5. Which age group has the highest churn?

```python
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[18,30,45,60,100],
    labels=['Young','Adult','Middle','Senior']
)
(df.groupby('AgeGroup' ,observed=False)['Exited'].mean()*100).round(2).sort_values(ascending=False)
```

Financial Analysis

Q6. What is the average balance of churned vs retained customers?

```python
df.groupby('Exited')['Balance'].mean().round(2)
```

07.Do customers with higher salaries churn more?

```python
df['SalaryGroup'] = pd.qcut(
    df['EstimatedSalary'],
    q=4,
    labels=['Low', 'Medium', 'High', 'Very High']
)
df.groupby('SalaryGroup',observed=False)['Exited'].mean() * 100
```

Quantile-based binning (qcut) was applied to segment customers into equal-sized salary groups. This approach helps ensure fair comparison across income levels by avoiding imbalanced group sizes, which can occur with fixed interval binning (cut).

08.Does credit score affect churn?

```python
df['Credit_band'] = pd.cut(
    df['CreditScore'],
    bins= [0,500, 600, 700, 800, 1000],
    labels=['Very Low','Low','Medium','High','Very High'])
(df.groupby('Credit_band' ,observed=False)['Exited'].mean()*100).round(2)
```

Customers with lower credit scores tend to have higher churn rates, suggesting that credit score may influence customer churn.

Customer Segmentation Analysis

Q9. Do active members churn less?

```python
(df.groupby('IsActiveMember')['Exited'].mean() * 100).round(2)
```

Q10. Which geography has the highest average balance among churned customers only?

```python
df[df['Exited'] == 1] \
  .groupby('Geography')['Balance'] \
  .mean() \
  .round(2) \
  .reset_index (name = 'avg balance') \
  .sort_values(by='avg balance',ascending=True)
```

This analysis identifies which geographic region has the highest average balance among customers who churned, helping highlight potential revenue loss by region.
